```python
df_with_indices = []
aux = []
for i in df_list:
    df = i[0]
    index_tuple = i[1]
    do_avg = i[2]
    # Assign the tuple as a multi-level index to the dataframe
    df.index = pd.MultiIndex.from_tuples([index_tuple], names=['Dataset', 'Horizon'])
    # Append the dataframe to the list
    df_with_indices.append(df)
    aux.append(df)
    if do_avg == True:
        aux = pd.DataFrame(pd.concat(aux)).groupby(level=0).mean()
        aux.index = pd.MultiIndex.from_product([aux.index, ['Avg']], names=['Dataset', 'Horizon'])
        df_with_indices.append(aux)
        aux = []
        
df = pd.concat(df_with_indices)
df.columns.name = "Metric"
df = df.rename(columns={
    "train_mse" : "MSE (Train)", "train_mae" : "MAE (Train)", 
    "val_mse" : "MSE (Valid)", "val_mae" : "MAE (Valid)", 
    "test_mse" : "MSE (Test)", "test_mae" : "MAE (Test)",
    "elapsed_time" : "Elapsed Time (s)", "vram_usage_gb": "VRAM Usage (Gb)"
    })
```

```python
# Function to apply bold to Avg rows
def highlight_avg(val):
    return ['font-weight: bold' if v[1] == 'Avg' else '' for v in val.index]

# Apply the style to the DataFrame
styled_df = df.style.background_gradient(cmap='YlGnBu').apply(highlight_avg, axis=0).set_table_styles(
    [
        {'selector': 'th', 'props': [('border', '2px solid black')]},  # Header borders
        {'selector': 'td', 'props': [('border', '1px solid gray')]},   # Cell borders
        {'selector': 'thead', 'props': [('border', '2px solid black')]},  # Outer header border
        {'selector': '.index_name', 'props': [('border', '2px solid black')]},  # Index border
        {'selector': 'tbody tr', 'props': [('border', '2px solid black')]},  # Row border
    ]
)

# Display the styled DataFrame (in a notebook or output to HTML)
styled_df
```

```python
df_with_indices = []
for i in df_list:
    df = i[0]
    index_tuple = i[1]
    # Assign the tuple as a multi-level index to the dataframe
    df.index = pd.MultiIndex.from_tuples([index_tuple], names=['Dataset', 'Horizon', 'Configuration'])
    # Append the dataframe to the list
    df_with_indices.append(df)

df = pd.concat(df_with_indices)
df.columns.name = "Metric"
df = df.rename(columns={
    "train_mse" : "MSE (Train)", "train_mae" : "MAE (Train)", 
    "val_mse" : "MSE (Valid)", "val_mae" : "MAE (Valid)", 
    "test_mse" : "MSE (Test)", "test_mae" : "MAE (Test)",
    "elapsed_time" : "Elapsed Time (s)", "vram_usage_gb": "VRAM Usage (Gb)"
    })
```

```python
def highlight_min(s):
    '''
    Highlight the minimum in a Series yellow.
    '''
    is_min = s == s.min()
    return ['background-color: yellow' if v else '' for v in is_min]

def style_horizon_group(group):
    """
    Apply highlighting to a group of rows that share the same Horizon value.
    """
    return group.style.background_gradient().apply(highlight_min, subset=["MSE (Valid)", "MAE (Valid)", "MSE (Test)", "MAE (Test)"]).set_table_styles(
    [
        {'selector': 'th', 'props': [('border', '2px solid black')]},  # Header borders
        {'selector': 'td', 'props': [('border', '1px solid gray')]},   # Cell borders
        {'selector': 'thead', 'props': [('border', '2px solid black')]},  # Outer header border
        {'selector': '.index_name', 'props': [('border', '2px solid black')]},  # Index border
        {'selector': 'tbody tr', 'props': [('border', '2px solid black')]},  # Row border
    ]
)

# Apply styling for each group and concatenate the results
styled_dfs = []
for horizon, group in df.groupby(level='Horizon'):
    styled_dfs.append(style_horizon_group(group))

# If you're in a Jupyter Notebook, you can display the first styled dataframe as a test
styled_df = styled_dfs[0]
```
